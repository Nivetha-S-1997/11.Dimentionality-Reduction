```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
dataset=pd.read_csv('Wine.csv')
dataset
```

```python
independent=dataset.drop('Customer_Segment',axis=1)
dependent=dataset['Customer_Segment']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.25,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
def PrincipleComponentAnalysis(X_train,X_test,n):
    from sklearn.decomposition import PCA
    pca=PCA(n_components=n,svd_solver='full')
    X_train=pca.fit_transform(X_train)
    X_test=pca.transform(X_test)
    explained_ratio=pca.explained_variance_ratio_
    return X_train,X_test,explained_ratio
```

```python
def r2_prediction(regressor,X_test,Y_test):
    Y_pred=regressor.predict(X_test)
    
    R2=r2_score(Y_test,Y_pred)
    return R2
```

```python
def linear(X_train,Y_train):
    regressor=LinearRegression()
    regressor.fit(X_train,Y_train)
    
    R2=r2_prediction(regressor,X_test,Y_test)
    return regressor,R2

def svrl(X_train,Y_train):
    regressor=SVR(kernel='linear')
    regressor.fit(X_train,Y_train)
    
    R2=r2_prediction(regressor,X_test,Y_test)
    return regressor,R2

def svrnl(X_train,Y_train):
    regressor=SVR(kernel='rbf')
    regressor.fit(X_train,Y_train)
    
    R2=r2_prediction(regressor,X_test,Y_test)
    return regressor,R2

def decisiontree(X_train,Y_train):
    regressor=DecisionTreeRegressor(random_state=0)
    regressor.fit(X_train,Y_train)
    
    R2=r2_prediction(regressor,X_test,Y_test)
    return regressor,R2

def randomforest(X_train,Y_train):
    regressor=RandomForestRegressor(n_estimators=10,random_state=0)
    regressor.fit(X_train,Y_train)
    
    R2=r2_prediction(regressor,X_test,Y_test)
    return regressor,R2
```

```python
def pca_table(acclin,accsvml,accsvmnl,accdt,accrf):
    dataframe=pd.DataFrame(index=['PCA'],columns=['Linear','SVML','SVMNL','DecisionTree','RandomForest'])

    for number,index in enumerate(dataframe.index):
        dataframe['Linear'][index]=acclin[number]
        dataframe['SVML'][index]=accsvml[number]
        dataframe['SVMNL'][index]=accsvmnl[number]
        dataframe['DecisionTree'][index]=accdt[number]
        dataframe['RandomForest'][index]=accrf[number]
    return dataframe
```

```python
acclin=[]
accsvml=[]
accsvmnl=[]
accdt=[]
accrf=[]
```

```python
X_train,X_test,explained_ratio=PrincipleComponentAnalysis(X_train,X_test,1)
explained_ratio
```

```python
regressor,r2_lin=linear(X_train,Y_train)
acclin.append(r2_lin)
            
regressor,r2_svrl=svrl(X_train,Y_train)
accsvml.append(r2_svrl)
            
regressor,r2_svrnl=svrnl(X_train,Y_train)
accsvmnl.append(r2_svrnl)
            
regressor,r2_dt=decisiontree(X_train,Y_train)
accdt.append(r2_dt)
            
regressor,r2_rf=randomforest(X_train,Y_train)
accrf.append(r2_rf)
```

```python
result=pca_table(acclin,accsvml,accsvmnl,accdt,accrf)
```
