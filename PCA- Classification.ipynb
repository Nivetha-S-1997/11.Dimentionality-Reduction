```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
dataset=pd.read_csv('Wine.csv')
dataset
```

```python
independent=dataset.drop('Customer_Segment',axis=1)
dependent=dataset['Customer_Segment']
```

```python
independent
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.25,random_state=0)
#X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
#X_train
```

```python
def PrincipleComponentAnalysis(X_train,X_test,n):
    from sklearn.decomposition import PCA
    pca=PCA(n_components=n,svd_solver='auto')
    X_train=pca.fit_transform(X_train)
    X_test=pca.transform(X_test)
    explained_ratio=pca.explained_variance_ratio_
    return X_train,X_test,explained_ratio
```

```python
def pred_confusion(classifier,X_test,Y_test):
    Y_pred=classifier.predict(X_test)
    
    cm=confusion_matrix(Y_test,Y_pred)
    accuracy=accuracy_score(Y_test,Y_pred)
    report=classification_report(Y_test,Y_pred)
    return cm,accuracy,report
```

```python
def logistic(X_train,Y_train):
    classifier=LogisticRegression(random_state=0)
    classifier.fit(X_train,Y_train)
    
    cm,accuracy,report=pred_confusion(classifier,X_test,Y_test)
    return classifier,cm,accuracy,report

def svml(X_train,Y_train):
    classifier=SVC(kernel='linear',random_state=0)
    classifier.fit(X_train,Y_train)
    
    cm,accuracy,report=pred_confusion(classifier,X_test,Y_test)
    return classifier,cm,accuracy,report

def svmnl(X_train,Y_train):
    classifier=SVC(kernel='rbf',random_state=0)
    classifier.fit(X_train,Y_train)
    
    cm,accuracy,report=pred_confusion(classifier,X_test,Y_test)
    return classifier,cm,accuracy,report

def decisiontree(X_train,Y_train):
    classifier=DecisionTreeClassifier(criterion='entropy',random_state=0)
    classifier.fit(X_train,Y_train)
    
    cm,accuracy,report=pred_confusion(classifier,X_test,Y_test)
    return classifier,cm,accuracy,report

def randomforest(X_train,Y_train):
    classifier=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
    classifier.fit(X_train,Y_train)
    
    cm,accuracy,report=pred_confusion(classifier,X_test,Y_test)
    return classifier,cm,accuracy,report

def naive_bayes(X_train,Y_train):
    classifier=GaussianNB()
    classifier.fit(X_train,Y_train)
    
    cm,accuracy,report=pred_confusion(classifier,X_test,Y_test)
    return classifier,cm,accuracy,report

def knn(X_train,Y_train):
    classifier=KNeighborsClassifier(n_neighbors=15,metric='minkowski',p=2)
    classifier.fit(X_train,Y_train)
    
    cm,accuracy,report=pred_confusion(classifier,X_test,Y_test)
    return classifier,cm,accuracy,report
```

```python
def pca_table(acclog,accsvml,accsvmnl,accdt,accrf,accnb,accknn):
    dataframe=pd.DataFrame(index=['PCA'],columns=['Logistic','SVML','SVMNL','DecisionTree','RandomForest','NaiveBayes','KNN'])

    for number,index in enumerate(dataframe.index):
        dataframe['Logistic'][index]=acclog[number]
        dataframe['SVML'][index]=accsvml[number]
        dataframe['SVMNL'][index]=accsvmnl[number]
        dataframe['DecisionTree'][index]=accdt[number]
        dataframe['RandomForest'][index]=accrf[number]
        dataframe['NaiveBayes'][index]=accnb[number]
        dataframe['KNN'][index]=accknn[number]
    return dataframe
```

```python
acclog=[]
accsvml=[]
accsvmnl=[]
accdt=[]
accrf=[]
accnb=[]
accknn=[]
```

```python
X_train,X_test,explained_ratio=PrincipleComponentAnalysis(X_train,X_test,1)
explained_ratio
```

```python
classifier,cm,accuracy,report=logistic(X_train,Y_train)
acclog.append(accuracy)
    
classifier,cm,accuracy,report=svml(X_train,Y_train)
accsvml.append(accuracy)
    
classifier,cm,accuracy,report=svmnl(X_train,Y_train)
accsvmnl.append(accuracy)
    
classifier,cm,accuracy,report=decisiontree(X_train,Y_train)
accdt.append(accuracy)
    
classifier,cm,accuracy,report=randomforest(X_train,Y_train)
accrf.append(accuracy)
    
classifier,cm,accuracy,report=naive_bayes(X_train,Y_train)
accnb.append(accuracy)
    
classifier,cm,accuracy,report=knn(X_train,Y_train)
accknn.append(accuracy)
```

```python
result=pca_table(acclog,accsvml,accsvmnl,accdt,accrf,accnb,accknn)
```
